```python
import pandas as pd
```

```python
data = pd.read_csv('covid_19_country_wise_latest.csv')
```

```python
data
```

```python
# ques1 
data.head()
```

```python
#ques2
data.set_index("Unnamed: 0",inplace=True)
```

```python
#ques3
data.index.name="index"
```

```python
#ques4
def replace_nonewithcolounm(data):
    data.columns=data.columns.str.replace(' ','_',regex=True)

    return data


data=replace_nonewithcolounm(data)
data
```

```python
#ques5
data.isna().sum()
data.columns
data.dtypes
```

```python
#ques6
data.isna().sum()
data.New_cases.fillna(data.New_cases.mean(),inplace=True)
data.New_deaths.fillna(data.New_deaths.mean(),inplace=True)
```

```python
#ques7
unique_countries=data['Country/Region'].nunique()
unique_countries
#No unique countries
```

```python
#ques 8
duplicate_country=data.duplicated(subset=['Country/Region'])
data[duplicate_country]
```

```python
#ques9
data.Confirmed.mean()
data.Confirmed.median()
data.Confirmed.std()
```

```python
#ques10
data['Deaths']=pd.to_numeric(data['Deaths'],errors="coerce")
```

```python
mean_deaths=data.Deaths.mean()
```

```python
data.fillna({"Deaths":mean_deaths},inplace=True)
```

```python
data.Deaths.info()
```

```python
#ques11
data.Deaths.dtypes
```

```python
#ques12
data.Deaths.sum()
```

```python
data.Recovered.sum()
```

```python
#ques13
data['total']=data.Confirmed+data.New_cases
data[data['total']>1_000_000]
data[data['total']>1_000_000].shape[0]
```

```python
#ques14
data['recovery_rate']=data.Recovered/data.total
data[data['recovery_rate']>=0.95].shape[0]
```

```python
#ques15
data.drop(columns=['WHO_Region','Confirmed'])
```

```python
#ques16
max_death=data['Country/Region'].max()
data[data['Country/Region']==max_death]
```

```python
#ques17
data.sort_values(by="Deaths",ascending=False)[['Country/Region','Deaths']]
```

```python
#ques18
data['Total_cases']=data.Deaths+data.Recovered+data.Active
data.sort_values(by="Total_cases")
```

```python
#ques19
data['Death_rate']=(data.Deaths/data.Total_cases)*100
```

```python
data.Death_rate.max()
```

```python
#ques20
high_cases_low_deaths=data[(data["Death_rate"] < 5) & (data["Total_cases"].diff() > 0)]["Country/Region"]
```

```python
high_cases_low_deaths.values
```
